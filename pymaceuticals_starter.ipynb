## Analysis
* Overall, Capomulin outperforms other drugs (Infubinol, Ketapril).
* Accorind to Figure "Tumor Response to Treatment" and "Tumor Change over 45 Day Treatment", Capomulin is the only drug that can reduce tumor volume.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse=pd.read_csv(mouse_drug_data_to_load)
clinical=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data=pd.merge(clinical,mouse,how='left',on="Mouse ID")

# Display the data table for preview
data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_group=data.groupby(["Drug","Timepoint"]).mean()['Tumor Volume (mm3)']

# Convert to DataFrame
tumor_mean=pd.DataFrame(tumor_group).reset_index()

# Preview DataFrame
tumor_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem=pd.DataFrame(data.groupby(["Drug","Timepoint"]).sem()['Tumor Volume (mm3)']).reset_index()


# Preview DataFrame
tumor_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_mean_reform=tumor_mean.pivot(index="Timepoint",columns="Drug")

#drop a level from a multi-level column index
tumor_mean_reform.columns = tumor_mean_reform.columns.droplevel()
# Preview that Reformatting worked
tumor_mean_reform.head()
```

```python
tumor_sem_reform=tumor_sem.pivot(index="Timepoint",columns="Drug")

#drop a level from a multi-level column index
tumor_sem_reform.columns = tumor_sem_reform.columns.droplevel()
# Preview that Reformatting worked
tumor_sem_reform.head()
```

```python
tumor_mean_reform.index
```

```python
# Generate the Plot (with Error Bars)

# set up legend parameters

colors={"Capomulin":["red","o"],
        "Infubinol":["blue","^"],
        "Ketapril":["green","s"],
        "Placebo":["black","D"]}

fig,ax=plt.subplots()

for drug in colors.keys():
    ax.errorbar(tumor_mean_reform.index, tumor_mean_reform[drug], tumor_sem_reform[drug],
                color=colors[drug][0],
                fmt=colors[drug][1],
                linestyle=":")
    
plt.legend(loc="best")
plt.grid()
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")

# Save the Figure
plt.savefig("images//Tumor_response_to_treatment.png")
plt.show()

```

##  Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_group=data.groupby(["Drug","Timepoint"]).mean()['Metastatic Sites']

# Convert to DataFrame
meta_mean=pd.DataFrame(meta_group)

# Preview DataFrame
meta_mean.head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_sem=pd.DataFrame(data.groupby(["Drug","Timepoint"]).sem()['Metastatic Sites'])

# Preview DataFrame
meta_sem.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_mean=meta_mean.reset_index()
meta_mean_reform=meta_mean.pivot(index="Timepoint",columns="Drug",values='Metastatic Sites')


# Preview that Reformatting worked
meta_mean_reform.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_sem=meta_sem.reset_index()
meta_sem_reform=meta_sem.pivot(index="Timepoint",columns="Drug",values='Metastatic Sites')


# Preview that Reformatting worked
meta_sem_reform.head()
```

```python
# Generate the Plot (with Error Bars)
fig,ax=plt.subplots()

for drug in colors.keys():
    ax.errorbar(meta_mean_reform.index, meta_mean_reform[drug], meta_sem_reform[drug],
                color=colors[drug][0],
                fmt=colors[drug][1],
               linestyle=":")
    
plt.legend(loc="best")
plt.grid()
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")

# Save the Figure
plt.savefig("images//Metastatic_Spread_during_treatment.png")
plt.show()


```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
sur_group=data.groupby(["Drug","Timepoint"]).count()['Mouse ID']

# Convert to DataFrame
survival=pd.DataFrame(sur_group).reset_index()
survival.rename(columns={"Mouse ID":"Mouse Count"},inplace=True)
# Preview DataFrame
survival.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_reform=survival.pivot(index="Timepoint",columns="Drug",values='Mouse Count')


# Preview that Reformatting worked
survival_reform.head()
```

```python
# Generate the Plot (Accounting for percentages)
target_survival=survival_reform[colors.keys()]

surv_per=target_survival.div(target_survival.iloc[0],axis='columns')*100

surv_per
```

```python
fig,ax=plt.subplots()

for drug in colors.keys():
    ax.plot(surv_per.index, surv_per[drug], 
            color=colors[drug][0],
            marker=colors[drug][1],
            linestyle=":")
    
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")
plt.legend(loc='best')
plt.grid()


# Save the Figure
plt.savefig("images//Survival_during_treatment.png")

# Show the Figure
plt.show()
```

```python
# Store all Relevant Percent Changes into a Tuple
per_change=tuple(drug_change[list(colors.keys())])

# Splice the data between passing and failing drugs
bins=[-100,0,100]
label_name=["pass","fail"]
labels=pd.cut(per_change,bins,labels=label_name)

# ref_link
#https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py
    

# Orient widths. Add labels, tick marks, etc.
ind=np.arange(len(per_change))
width=1

fig,ax=plt.subplots()
recsf=ax.bar(ind[0],per_change[0],width,color="green",ec='white')
recsp=ax.bar(ind[1:],per_change[1:],width,color="red",ec='white')

ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Day Treatment")
ax.set_xticks(ind)
ax.set_xticklabels(tuple(colors.keys()))
ax.set_ylim(-26,71)
ax.grid(True)

# Use functions to label the percentages of changes
def autolabel_pass(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('%d%%' % int(height),
                    xy=(rect.get_x() + rect.get_width()/2,height/12),color='white',
                    ha="center", va='bottom')
        
        
def autolabel_fail(rects, xpos='center'):
    
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('-%d%%' % int(height),
                    xy=(rect.get_x() + rect.get_width()/2,-8), color='white',
                    ha="center", va='bottom')



# Call functions to implement the function calls
autolabel_pass(recsp)
autolabel_fail(recsf)

fig.tight_layout()



# Save the Figure
plt.savefig("images//Tumor_change_over45_Day.png")

# Show the Figure
plt.show()
```
